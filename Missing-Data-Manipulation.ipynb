
Cette fonction check_null_percentage(df) calcule le pourcentage de valeurs nulles dans chaque colonne du DataFrame df. Elle renvoie un DataFrame contenant le nom de la colonne et le pourcentage de valeurs nulles pour chaque colonne.

```python
def check_null_percentage(df):

    total_rows = len(df)
    null_percentages = (df.isnull().sum() / total_rows) * 100
    df_null_percentage = pd.DataFrame({'Column': null_percentages, 'Null_per': null_percentages.values})

    return df_null_percentage
```

```python
null_percentage = check_null_percentage(df)
```

```python
null_percentage
```

```python
null_more_than_70 = null_percentage[null_percentage['Null_per']>=70]
```

remplir les données manquantes
* on va remplacer les valeurs de type categorique avec Na pour l'encoder apres avec les autres valeurs comme le cas ou il ya l'abscence de chose
* on va remplacer les valeur float de LotFrontage et MasVnrArea par leurs moyennes
* on va remplacer la valeur de MasVnrType par le mode car c'est une années et one ne peut pas la remplacer par sa moyenne pour avoir une valeur entiere ou par 'na' pour garder le type 'int'

```python
df["GarageYrBlt"].unique()
```

```python
fill_with_NA=[ 'MasVnrType', 'FireplaceQu','Electrical','GarageType','GarageFinish', 'GarageQual', 'GarageCond', 'BsmtExposure', 'BsmtFinType2', 'BsmtFinType1', 'BsmtCond', 'BsmtQual']
fill_with_mean=['LotFrontage', 'MasVnrArea']
fill_with_mode=["GarageYrBlt"]

for i in fill_with_NA:
  df[i]=df[i].fillna("NA")

for i in fill_with_mean:
  df[i]=df[i].fillna(df[i].mean())

for i in fill_with_mode:
  df[i]=df[i].fillna(df[i].value_counts().idxmax())
```

encoder les donnees categoriques

```python
from sklearn.preprocessing import LabelEncoder

non_numeric_cols = df.select_dtypes(exclude=['number']).columns
label_encoder = LabelEncoder()

for col in non_numeric_cols:
      df[col] = label_encoder.fit_transform(df[col])
```
